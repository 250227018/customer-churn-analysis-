```python


# Example: df is your dataframe
# df = pd.read_csv("your_data.csv")

# Bin continuous variables so countplot makes sense
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 25, 40, 60, 100],
                         labels=['<25', '25–40', '40–60', '60+'])

df['MonthlyChargesGroup'] = pd.cut(df['MonthlyCharges'], bins=4)
df['TotalChargesGroup'] = pd.cut(df['TotalCharges'], bins=4)

# Columns to plot
plot_cols = [
    'Gender', 'AgeGroup',  'ContractType',
    'InternetService', 'TechSupport',
    'MonthlyChargesGroup', 'TotalChargesGroup', 'Churn'
]

plt.figure(figsize=(20,18))

for i, col in enumerate(plot_cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=df, x=col, hue='Churn')
    plt.xticks(rotation=45)
    plt.title(f'Countplot of {col}')

plt.tight_layout()
plt.show()
```

Explanation of each plot (2 lines each)
Gender
Shows the number of male and female customers and how churn differs by gender.
Helps check if churn is biased toward a specific gender.

AgeGroup
Displays customer distribution across age ranges and their churn behavior.
Useful for identifying age groups with higher churn risk.

Tenure
Shows how long customers stay with the company before churning.
Short-tenure customers usually show higher churn rates.

ContractType
Compares churn across month-to-month, yearly, and longer contracts.
Month-to-month contracts typically have the highest churn.


InternetService
Shows churn differences based on internet service type.
Certain services may be linked to dissatisfaction and higher churn.

TechSupport
Compares churn between customers with and without tech support.
Lack of tech support often correlates with higher churn.


MonthlyChargesGroup
Displays churn patterns across different monthly charge ranges.
Higher charges often lead to increased churn sensitivity.

TotalChargesGroup
Shows customer distribution based on total spending.
Lower total charges often indicate newer customers with higher churn.


Churn
Shows the overall balance between churned and retained customers.
Helps identify class imbalance in the dataset
